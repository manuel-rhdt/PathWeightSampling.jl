```julia
using PWS
using Plots
using LaTeXStrings
```

```julia
system = System()
```

```julia
p = plot([corr_ss(system) corr_xs(system) corr_sx(system) corr_xx(system)], 0, 600, 
    labels=[L"C_{ss}(t\prime-t)" L"C_{xs}(t\prime-t)" L"C_{sx}(t\prime-t)" L"C_{xx}(t\prime-t)"], 
    palette=:tab10, 
    xlabel=L"t\prime-t", 
    ylabel="correlation",
    size=(600, 200),
    linestyle=[:solid :dashdot :dashdotdot :solid],
    linewidth=2)
```

```julia
savefig(p, "correlation.svg")
```
